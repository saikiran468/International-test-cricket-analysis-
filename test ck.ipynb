```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df["players"],df["runs"])
plt.title("Test Runs by Players")
plt.xlabel("Players")
plt.ylabel("Runs")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(df["players"],df["average"])
plt.title("Batting Average")
plt.xlabel("Players")
plt.ylabel("Average")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(df["players"],df["balls"])
plt.title("tast stats")
plt.xlabel("players")
plt.ylabel("balls")
plt.xticks(rotation=45)
plt.show()
```
